```python
import pandas as pd
import numpy as np
import httpx
```

```python
boardid, secid = 'TQOB', 'SU26227RMFS7'
url_data = f"https://iss.moex.com/iss/engines/stock/markets/bonds/boards/{boardid}/securities/{secid}.json?iss.meta=off"
client = httpx.Client(http2=True)
response_moex = client.get(url=url_data)
bonds_data = response_moex.json()
bonds_data
```

```python
columns = bonds_data['securities'].get('columns')
data = pd.Series(data=bonds_data['securities'].get('data')[0], index=columns)
data
```
