```python
import pandas as pd
import numpy as np
from bonds import Bond

bond = Bond()
```

```python
# Обновление списка secid
resp_sec = bond.get_secid_bonds()
```

```python
all_data = []

if resp_sec:
    for k, v in resp_sec.items():
        for item in v:
            resp = bond.get_bond_data(boardid=k, secid=item)
            all_data.append(resp) if resp else None
```

```python
view = pd.DataFrame(all_data)
view.sort_values(by='cur_yields', ascending=False)[:15]
```
